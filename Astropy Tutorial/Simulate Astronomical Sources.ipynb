# How to Create Mock/ Fake/ Simulated Astronomical Sources: Stars/ Galaxies

Step 1: Import the libraries

```python
import numpy as np
import matplotlib.pyplot as plt
from astropy.modeling.functional_models import Gaussian2D, Sersic2D
```

Step 2: Create canvas: image size & pixel size

```python
#def image size
size = 300
# create @D meshgrid
x, y = np.meshgrid(np.arange(300), np.arange(300))
```

Step 3: Create Star image with Gaussian 2D model

```python
star = Gaussian2D(amplitude=5, x_mean=200, y_mean=200, x_stddev=5, y_stddev=5)
# Create a Star Image from model
star_image = star(x, y)
star_image
```

Step 4: Create a Galaxy image with Sersic Model

```python
galaxy = Sersic2D(amplitude=0.5, r_eff=50, n=3, x_0=100, y_0=100, theta=np.deg2rad(45))
```

```python
# Create galaxy image data as 2D
galaxy_image = galaxy(x, y)
galaxy_image
```

Step 5: Combine star + galaxy images

```python
mock_image = star_image + galaxy_image
mock_image
```

Step 6: Visualize mock image

```python
plt.imshow(mock_image, origin='lower', cmap='gray', vmin=)
```
